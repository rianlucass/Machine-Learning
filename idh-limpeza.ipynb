```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_idh = pd.read_csv('dados/Human Development Index - Full.csv')
```

```python
df_idh.head(10)
```

```python
df_idh.columns
```

```python
cols_hdi = ['Country'] + [col for col in df_idh.columns if col.startswith('Human Development Index')]
df_idh_clean = df_idh[cols_hdi]
```

```python
df_idh_clean.head(100)
```

```python
df_idh_clean.shape
```

```python
df_idh_clean.isnull().sum()
```

```python
df_idh_tratado = df_idh_clean.copy()
```

```python
for col in df_idh_tratado.columns[1:]:
    df_idh_tratado[col] = pd.to_numeric(df_idh_tratado[col], errors='coerce')
df_idh_tratado.iloc[:, 1:] = df_idh_tratado.iloc[:, 1:].interpolate(axis=1)
```

```python
df_idh_tratado.isnull().sum()
```

```python
df_idh_tratado.head(30)
```

```python
idh_com_nulos = df_idh_tratado[df_idh_tratado.isnull().any(axis=1)]
idh_com_nulos[['Country']]
```

```python
df_idh_tratado = df_idh_tratado.dropna()
```

```python
df_idh_tratado.head(50)
```

```python
df_idh_tratado.isnull().sum()
```

```python
df_idh_tratado.to_csv('dados/Human-Development-Index-concluido.csv', index=False)
```
