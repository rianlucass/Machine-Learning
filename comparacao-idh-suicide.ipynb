```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
idh_df = pd.read_csv("dados/Human-Development-Index-concluido.csv")
suicidio_df = pd.read_csv("dados/suicide_statistic_concluido_corrigido.csv")
```

```python
paises_disponiveis = sorted(suicidio_df['country'].unique())
```

```python
print(paises_disponiveis)
```

```python
paises = ["Germany", "Switzerland", "Mexico", "Brazil", "Kiribati", "Sao Tome and Principe"]
```

```python
idh_df['Country'] = idh_df['Country'].str.strip()
idh_filtrado = idh_df[idh_df['Country'].isin(paises)][['Country', 'Human Development Index (2021)']]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(idh_filtrado['Country'], idh_filtrado['Human Development Index (2021)'], color='green')
plt.title('IDH (2021) por País')
plt.xlabel('País')
plt.ylabel('IDH (2021)')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
suicidio_df['country'] = suicidio_df['country'].str.strip()
```

```python
suicidio_df = suicidio_df[suicidio_df['country'].isin(paises)]
suicidio_df['taxa_suicidio'] = (suicidio_df['suicides_no'] / suicidio_df['population']) * 100000
taxa_media = suicidio_df.groupby('country')['taxa_suicidio'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(taxa_media['country'], taxa_media['taxa_suicidio'], color='orange')
plt.title('Taxa Média de Suicídio por País (por 100 mil habitantes)')
plt.xlabel('País')
plt.ylabel('Taxa média de suicídio')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
